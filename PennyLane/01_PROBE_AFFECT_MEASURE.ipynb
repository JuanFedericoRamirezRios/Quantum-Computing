```python
import pennylane as qml
from pennylane import numpy as np
import matplotlib.pyplot as pp
```

```python
dev = qml.device('default.qubit', wires=1, shots=100)
```

```python
@qml.qnode(dev)
def Circuit(measured, input):
    qml.AmplitudeEmbedding(features=input,wires=0,normalize=True) #Normalize so that <psi|psi>=1
    qml.Hadamard(wires=0)
    if measured:
        qml.measure(wires=0)
    qml.Hadamard(wires=0)

    return qml.sample(wires = 0), qml.probs(wires = 0)


    
```

```python
qml.drawer.use_style("pennylane")
initState = np.array([2, 0])
```

```python
measure = False
qml.draw_mpl(Circuit)(measured = measure, input = initState, shots = None) # Avoid the use of device by none shots.
print('Circuit:')
pp.show()
print('Out put:') 
Circuit(measured = measure, input = initState)
```

```python
measure = True
fig, ax = qml.draw_mpl(Circuit)(measured = measure, input = initState, shots = None) # Avoid the use of device by none shots.
print('Circuit:')
pp.show()
print('Out put:') 
Circuit(measured = measure, input = initState)
```
