```python
from qiskit import QuantumCircuit
from qiskit.providers.basic_provider import BasicSimulator
simulator = BasicSimulator()
```

```python
def Circuit(measure, input):
    qc = QuantumCircuit(1,1)
    if input == 1:
        qc.x(0)
    qc.h(0)
    if measure:
        qc.measure(0,0)
    qc.h(0)
    qc.measure(0,0)
    return qc

# qc = QuantumCircuit(1,1)
# qc.h(0)
# qc.h(0)
# qc.measure(0,0)
# qc.draw(output='mpl')
```

```python
measure = False
input = 1
circuit = Circuit(measure, input)
circuit.draw(output='mpl')
```

```python
result = simulator.run(circuit,shots=100).result()
print(result.get_counts())
```

```python
measure = True
input = 0
circuit = Circuit(measure, input)
print(circuit.draw())
result = simulator.run(circuit,shots=100).result()
print('Results:', result.get_counts())
```
