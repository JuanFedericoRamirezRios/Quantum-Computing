```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.quantum_info import Statevector
from qiskit.providers.basic_provider import BasicSimulator
import numpy as np
simulator = BasicSimulator()
```

```python
def Circuit():
    a1 = QuantumRegister(1, "A1") # In (1, "A1") 1 is the number of qubits and "A1" is a name.
    a2 = QuantumRegister(1, "A2") 
    b = QuantumRegister(1, "B")
    qc = QuantumCircuit(a1, a2, b, ClassicalRegister(1), ClassicalRegister(1), ClassicalRegister(1))
    state = Statevector(np.array([(0.7)**0.5, (0.3)**0.5]))
    qc.initialize(state, a1) # In this case we are initializing the first qubit.
    qc.h(a2)
    qc.cx(a2, b)
    qc.cx(a1, a2)
    qc.h(a1)
    qc.measure(a1, 0)
    qc.measure(a2, 1)
    qc.cx(a2, b)
    qc.cz(a1,b)
    qc.measure(b, 2)
    
    return qc


Circuit().draw(output='mpl')
```
